```python
import re
import codecs
import json
from os.path import join
import os
import numpy as np
import multiprocessing
import pandas as pd
import csv
from datetime import datetime
from utils.cache import LMDBClient
from utils import data_utils
from utils import settings
import time
start_time = datetime.now()

################# Load and Save Data ################

def load_json(rfdir, rfname):
    with codecs.open(join(rfdir, rfname), 'r', encoding='utf-8') as rf:
        return json.load(rf)


def dump_json(obj, wfpath, wfname, indent=None):
    with codecs.open(join(wfpath, wfname), 'w', encoding='utf-8') as wf:
        json.dump(obj, wf, ensure_ascii=False, indent=indent)

```

```python
def dump_single_name_tojson(name_pubs):
    print('Run task (%s)...' % (os.getpid()))
    start1 = time.time()
    cnt = 0
    lc = LMDBClient('sci_all_data')
    
    for n,name in enumerate(name_pubs):
        pubs = name_pubs[name]##存储某一人名下的所有文章
#         print(n,name,len(pubs))  
  
        ##保存关系
        ###############################################################
        name_pubs_raw = {}
        for i,pid in enumerate(pubs):
            paper = lc.get(pid[0])
            paper['org_name'] = pid[1]
            paper.pop('abstract')
            paper.pop('uid')
            name_pubs_raw[pid[0]] = paper
        dump_json(name_pubs_raw, '/home/wss/data/disk1/sci_disamb_name_50', name+'.json', indent=4)
        cnt+=1
        print('写入单个同名论文集数量: %d'%(cnt))
    print('task %s run %0.2f seconds.' % (os.getpid(), (time.time() - start1)))
        ###############################################################

def chunks(data, SIZE=1000):
    for i in range(0, len(data), SIZE):
        yield dict(list(data.items())[i:i+SIZE])

##加载需要消歧的数据
name_pubs = load_json("data","sci_author_data_50_all.json")
print('数据集同名数量：', len(name_pubs))

if __name__ =='__main__':
    processor = 10
    p = multiprocessing.Pool(processor)
    for i,item in enumerate(chunks(name_pubs,10000)):
#         name_pub =item
        if i < 10:
            p.apply_async(dump_single_name_tojson, args=(item,))
    p.close()
    p.join()
    print('将同名论文集存储为单个json文件 总用时 ：',datetime.now()-start_time)
```
